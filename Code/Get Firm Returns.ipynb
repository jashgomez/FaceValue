```python
import yfinance as yf
import pandas as pd
from tqdm import tqdm

# Re-defining firm_df with correct headers
dir = "/content/drive/MyDrive/Colab Notebooks/Axion FIN 377 Final Project/Photos Handling Code/" # Ensure dir is available
firm_df = pd.read_csv(dir+"CEO_Data_Correct.csv") # Let pandas infer header
firm_df['Year'] = firm_df['Year'].astype(int) # Ensure Year is integer type

# Function to check if a ticker is valid
def is_ticker_valid(ticker, start_date='2010-01-01', end_date='2019-12-31'):
    try:
        data = yf.Ticker(ticker).history(start=start_date, end=end_date)
        return not data.empty
    except Exception:
        return False

# Extract tickers
tickers = firm_df['Ticker'].drop_duplicates().tolist()
print(tickers)

# Validate tickers with progress bar
valid_tickers = [ticker for ticker in tqdm(tickers, desc="Validating tickers") if is_ticker_valid(ticker)]

# Download data for all valid tickers at once
if valid_tickers:
    all_data = yf.download(valid_tickers, start='2010-01-01', end='2019-12-31', group_by='ticker')
else:
    print("No valid tickers found.")
    all_data = None
```

```python
all_data
```

```python
import pandas as pd

# Assume 'all_data' is your DataFrame

# Initialize an empty list to collect mini-DataFrames
df_list = []

# Loop through each ticker in the first level of columns
for ticker in valid_tickers:
    try:
        # Get the Close prices for this ticker
        close_prices = all_data[ticker]['Close']

        # Calculate daily returns
        returns = close_prices.pct_change()

        # Build mini DataFrame
        temp_df = pd.DataFrame({
            'Ticker': ticker,
            'Year': close_prices.index.year,
            'Date': close_prices.index,
            'Close': close_prices.values,
            'Return': returns.values
        })

        df_list.append(temp_df)

    except Exception as e:
        print(f"Error processing {ticker}: {e}")

# Combine all into one big DataFrame
stock_prices = pd.concat(df_list, ignore_index=True)
```

```python
firm_returns = pd.merge(stock_prices, firm_df, how='left', on = ['Ticker', 'Year'])
```

```python
firm_returns
```

```python
# Cumulative Returns on a daily basis
```

```python
firm_returns['Tenure_Cum_Ret_Daily'] = \
(
    firm_returns
    .assign(R=1+stock_prices['Return'])
    .groupby(['Ticker', 'CEO'])
    ['R'].cumprod()
)
```

```python
firm_returns['Year_Cum_Ret_Daily'] = \
(
    firm_returns
    .assign(R=1+stock_prices['Return'])
    .groupby(['Ticker', 'Year'])
    ['R'].cumprod()
)
```

```python
# Cumulative Returns on a fixed basis
```

```python
firm_returns['Tenure_Cum_Ret_Overall'] = (
    (1 + stock_prices['Return'])  # no need to .assign
    .groupby([firm_returns['Ticker'], firm_returns['CEO']])
    .transform('prod')  # broadcast full tenure cumulative return to each row
)
```

```python
firm_returns['Year_Cum_Ret_Overall'] = (
    (1 + stock_prices['Return'])  # no need to .assign
    .groupby([firm_returns['Ticker'], firm_returns['Year']])
    .transform('prod')  # broadcast full tenure cumulative return to each row
)
```

```python
ceo_year_returns = firm_returns.groupby(['CEO', 'Year', 'Ticker', 'Company']).agg(
    mean_daily_return=('Return', 'mean'),
    final_year_cum_ret_daily=('Year_Cum_Ret_Daily', 'last'),
    final_tenure_cum_ret_daily=('Tenure_Cum_Ret_Daily', 'last'),
    year_cum_ret_overall=('Year_Cum_Ret_Overall', 'first'), # These are already year-fixed
    tenure_cum_ret_overall=('Tenure_Cum_Ret_Overall', 'first') # These are already tenure-fixed
).reset_index()

ceo_year_returns
```
